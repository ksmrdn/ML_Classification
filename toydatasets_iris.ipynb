```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
```

```python
i = load_iris()
dir(i)
```

```python
i['feature_names']
# df = pd.DataFrame(i['data'], columns=i['feature_names'])
df = pd.DataFrame(i['data'], columns=['SL','SW','PL','PW'])
df.head()
```

```python
i['target_names']
```

```python
df['spesies'] = i['target']
df['nama spesies'] = df['spesies'].apply(lambda x: i['target_names'][x])
# df.drop(['namaSP'], axis=1)
```

```python
import matplotlib.pyplot as plt

df0 = df[df['spesies'] == 0]
df1 = df[df['spesies'] == 1]
df2 = df[df['spesies'] == 2]

plt.figure(figsize=(12,8))
plt.subplot(121)
plt.plot(df0['SL'], df0['SW'], 'ro')
plt.plot(df1['SL'], df1['SW'], 'go')
plt.plot(df2['SL'], df2['SW'], 'bo')
plt.legend(i['target_names'])
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('length (cm)')
plt.ylabel('width (cm)')

plt.subplot(122)
plt.plot(df0['PL'], df0['PW'], "ro")
plt.plot(df1['PL'], df1['PW'], 'go')
plt.plot(df2['PL'], df2['PW'], 'bo')
plt.legend(i['target_names'])
plt.title('Petal Length vs Petal Width')
plt.xlabel('length (cm)')
plt.ylabel('width (cm)')

plt.suptitle('IRIS DATA SET')
```
