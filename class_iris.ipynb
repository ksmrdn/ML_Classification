```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### 1. Preparing Dataset

```python
from sklearn.datasets import load_iris
dataIris = load_iris()
#dir(dataIris)
df = pd.DataFrame(dataIris['data'],
    columns = dataIris['feature_names'])
df
```

```python
df['target'] = dataIris['target']
df['spesies'] = df['target'].apply(lambda x: dataIris['target_names'][x])
df
```

```python
import seaborn as sb
sbH = df.corr()
sb.heatmap(sbH, annot=True)
plt.ylim(7,-0.5)
```

### 2. Splitting Dataset

```python
from sklearn.model_selection import train_test_split
```

```python
xtr, xts, ytr, yts = train_test_split(df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']], df['target'])
```

```python
xtr
```

```python
df = df.rename(columns={'sepal length (cm)': 'SL', 'sepal width (cm)': 'SW', 'petal length (cm)': 'PL', 'petal width (cm)': 'PW'})
df
```

```python
plt.figure()
plt.subplot(121)
plt.plot(
    df['SL'][df['target'] == 0], df['SW'][df['target']==0], 'r.')
plt.plot(
    df['SL'][df['target'] == 1], df['SW'][df['target']==1], 'g.')
plt.plot(
    df['SL'][df['target'] == 2], df['SW'][df['target']==2], 'b.')

plt.subplot(122)
plt.plot(
    df['PL'][df['target'] == 0], df['PW'][df['target']==0], 'r.')
plt.plot(
    df['PL'][df['target'] == 1], df['PW'][df['target']==1], 'g.')
plt.plot(
    df['PL'][df['target'] == 2], df['PW'][df['target']==2], 'b.')
```

### 3. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lrA = LogisticRegression(solver='lbfgs', multi_class='auto')
lrB = LogisticRegression(solver='liblinear', multi_class='auto')
```

```python
lrA.fit(xtr, ytr)
lrB.fit(xtr, ytr)
```

### ROC AUC

```python
from sklearn.metrics import roc_auc_score, roc_curve, auc
```

```python
lrA.predict_proba([xts.iloc[0]])
```

```python
xts.iloc[0]
```

```python
yts.iloc[0]
```

```python
fpra0, tpra0, tha0 = roc_curve(yts, lrA.predict_proba(xts)[:,0], pos_label=0)
fprb0, tprb0, thb0 = roc_curve(yts, lrB.predict_proba(xts)[:,0], pos_label=0)
```

```python
fpra0, tpra0, tha0
```

```python
fprb0, tprb0, thb0
```

```python
fpra1, tpra1, tha1 = roc_curve(yts, lrA.predict_proba(xts)[:,1], pos_label=1)
fprb1, tprb1, thb1 = roc_curve(yts, lrB.predict_proba(xts)[:,1], pos_label=1)
```

```python
fpra2, tpra2, tha2 = roc_curve(yts, lrA.predict_proba(xts)[:,2], pos_label=2)
fprb2, tprb2, thb2 = roc_curve(yts, lrB.predict_proba(xts)[:,2], pos_label=2)
```

```python
aucA0 = roc_auc_score(np.array(yts.tolist()), lrA.predict_proba(xts), labels=[0,1,2], multi_class='ovo', average='macro')
aucB0 = roc_auc_score(np.array(yts.tolist()), lrB.predict_proba(xts), labels=[0,1,2], multi_class='ovo', average='macro')
aucA0, aucB0
```

```python
auca0 = auc(fpra0, tpra0)
aucb0 = auc(fprb0, tprb0)
auca1 = auc(fpra1, tpra1)
aucb1 = auc(fprb1, tprb1)
auca2 = auc(fpra2, tpra2)
aucb2 = auc(fprb2, tprb2)
auca0, aucb0, auca1, aucb1, auca2, aucb2
```
