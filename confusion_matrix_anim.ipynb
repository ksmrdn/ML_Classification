```python
import numpy as np
import pandas as pd
```

```python
y = ['Kucing', 'Anjing', 'Gajah', 'Kucing', 'Anjing', 'Gajah']
yp = ['Kucing', 'Kucing', 'Gajah', 'Kucing', 'Kucing', 'Gajah']
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y, yp, labels=['Kucing','Anjing','Gajah'])
cm
```

```python
from sklearn.metrics import plot_confusion_matrix as pcm
```

```python
dfCon = pd.DataFrame(cm, columns=['Kucing','Anjing','Gajah'],
                            index=['Kucing','Anjing','Gajah'])
dfCon
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
sb.heatmap(dfCon, annot=True, fmt='d')
plt.ylim(3,-0.05)
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, balanced_accuracy_score, f1_score, classification_report
```
