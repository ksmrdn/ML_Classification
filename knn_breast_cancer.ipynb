```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler 
from sklearn import decomposition
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
```

```python
pd.options.display.max_columns = 999
pd.options.display.max_rows = 999
```

```python
bc = load_breast_cancer()
dir(bc)
```

```python
dfB = pd.DataFrame(
    bc['data'], columns = bc['feature_names']
)
dfB
```

```python
dfB['target'] = bc['target']
dfB
```

```python
dfB.isnull().sum()
```

```python
dfB.corr().target.sort_values()
```

```python
print(dfB.corr()['target'][dfB.corr()['target'] < -0.5].index)
print(len(dfB.corr()['target'][dfB.corr()['target'] < -0.5].index))
```

```python
choosen = np.array(dfB.corr()['target'][dfB.corr()['target'] < -0.5].index)
```

```python
dfC = dfB[choosen]
dfC
```

```python
dfC['target'] = dfB['target']
dfC
```

```python
ss = StandardScaler()
```

```python
std = ss.fit_transform(dfC.iloc[:,:-1])
```

```python
dfStd = pd.DataFrame(std)
dfStd['target'] =  dfC['target']
dfStd
```

```python
#Check outlier, -2.5 < z-score > 2.5
Std_atas = dfStd[dfStd<2.5]
Std_atas
```

```python
outliers = Std_atas[Std_atas>-2.5]
outliers
```

```python
outliers.isnull().sum()
```

```python
inliers = outliers.dropna()
inliers.shape
```

```python
inliers
```

```python
y = np.array(inliers['target'])
```

### PCA

```python
datastd = np.array(inliers.iloc[:,:-1])
datastd
```

```python
pca = decomposition.PCA(n_components=3)
pca_n = pca.fit_transform(datastd)
pca_n[0]
```

```python
pca.n_components_
```

```python
pca.explained_variance_ratio_
```

```python
pca.explained_variance_ratio_.sum()
```

```python
#invers ke value semua
ss.inverse_transform(pca.inverse_transform(pca_n[0]))
```

```python
inliers[['target']]
```

```python
dfPca = pd.DataFrame(pca_n)
dfPca['class'] = y
dfPca
```
